```
police_shooting_datasets = {
    'Portland_OR': 'https://public.tableau.com/views/PPBOpenDataDownloads/OIS-All.csv?:showVizHome=no'
}
```

```
df = pd.read_csv(police_shooting_datasets['Portland_OR'], parse_dates=['Day of Date Time'])
```
